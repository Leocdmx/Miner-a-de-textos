## Tarea
Identificar correctamente todas las diferentes variantes de fecha codificadas en este conjunto de datos, normalizar y ordenar adecuadamente las fechas. Cada línea del archivo dates.txt corresponde a una nota médica. Cada nota tiene una fecha que debe extraerse, pero cada fecha está codificada en uno de muchos formatos. Aquí hay una lista de algunas de las variantes que puede encontrar en este conjunto de datos:

* 04/20/2009; 04/20/09; 4/20/09; 4/3/09
* Mar-20-2009; Mar 20, 2009; March 20, 2009; Mar. 20, 2009; Mar 20 2009;
* 20 Mar 2009; 20 March 2009; 20 Mar. 2009; 20 March, 2009
* Mar 20th, 2009; Mar 21st, 2009; Mar 22nd, 2009
* Feb 2009; Sep 2009; Oct 2010
* 6/2008; 12/2009
* 2009; 2010

Una vez que haya extraído estos patrones de fecha del texto, el siguiente paso es clasificarlos en orden cronológico ascendente de acuerdo con las siguientes reglas:

* Todas las fechas están en formato xx/xx/ xx son mm/dd/aa
* Todas las fechas en las que el año está codificado en solo dos dígitos corresponden a años posteriores a la década de 1900 (p. Ej., 1/5/89 es el 5 de enero de 1989).
* Si falta el día (p. Ej., 9/2009), suponga que es el primer día del mes (p. Ej., septiembre, 1 de 2009).
* Si falta el mes (por ejemplo, 2010), suponga que es el primero de enero de ese año (p. Ej., enero, 1 de 2010).
* Tenga cuidado con los posibles errores tipográficos, ya que este es un conjunto de datos derivados de la vida real.

```python
# Función para normalizar fechas en formato datetime



def fechas_ordenadas():
    #librerias
    import re
    from datetime import datetime

    #Carga del texto
    with open("dates.txt", "r", encoding="utf-8") as file:
        lineas = file.readlines()


    #Patron de busqueda
    patron = r'\d{1,2}(?:\sde)?\s[a-zA-Z]{3,9}(?:\sde)?\s(?:\d{4}|\d{2})|[a-zA-Z]{3,9}(?:.)?[\s][\d]{2}[,]?[\s][\d]{2,4}|\d{1,2}[-\s/](?:0[1-9]|[12][0-9]|3[01])[-\s/](?:\d{4}|\d{2})|[a-zA-Z]{3,9}(?:,)?[\s][\d]{4}|\d{1,2}[/]\d{2,4}|[1|2]\d{3}'


    Lista_fechas = []

    for linea in lineas:
        numeros = re.findall(patron, linea)
        Lista_fechas.append(numeros[0])

    lista_modificada = [texto.replace("-", "/") for texto in Lista_fechas]
    lista_modificada = [texto.replace("since ", "") for texto in lista_modificada]
    lista_modificada = [texto.replace("sSince ", "") for texto in lista_modificada]

    lista_modificada = [texto.replace("quit ", "") for texto in lista_modificada]
    lista_modificada = [texto.replace("Age, 19, ", "") for texto in lista_modificada]
    lista_modificada = [texto.replace("attempt ", "") for texto in lista_modificada]
    lista_modificada = [texto.replace("delivery ", "") for texto in lista_modificada]
    lista_modificada = [texto.replace("phrectomy ", "") for texto in lista_modificada]
    lista_modificada = [texto.replace("Age 16, ", "") for texto in lista_modificada]

    lista_modificada = [texto.replace("pOct", "Oct") for texto in lista_modificada]
    lista_modificada = [texto.replace("lNovember", "Nov") for texto in lista_modificada]
    lista_modificada = [texto.replace("IAug", "Aug") for texto in lista_modificada]
    lista_modificada = [texto.replace("yAug", "Aug") for texto in lista_modificada]
    lista_modificada = [texto.replace("AFeb", "Feb") for texto in lista_modificada]
    lista_modificada = [texto.replace("sSep", "Sep") for texto in lista_modificada]
    lista_modificada = [texto.replace("sNov", "Nov") for texto in lista_modificada]


    lista_modificada = [texto.replace("August", "Aug") for texto in lista_modificada]
    lista_modificada = [texto.replace("March", "Mar") for texto in lista_modificada]
    lista_modificada = [texto.replace("February", "Feb") for texto in lista_modificada]
    lista_modificada = [texto.replace("June", "Jun") for texto in lista_modificada]
    lista_modificada = [texto.replace("July", "Jul") for texto in lista_modificada]
    lista_modificada = [texto.replace("December", "Dec") for texto in lista_modificada]
    lista_modificada = [texto.replace("October", "Oct") for texto in lista_modificada]
    lista_modificada = [texto.replace("September", "Sep") for texto in lista_modificada]
    lista_modificada = [texto.replace("November", "Nov") for texto in lista_modificada]
    lista_modificada = [texto.replace("January", "Jan") for texto in lista_modificada]
    lista_modificada = [texto.replace("April", "Apr") for texto in lista_modificada]
    lista_modificada = [texto.replace("May", "May") for texto in lista_modificada]
    lista_modificada = [texto.replace("Janaury", "Jan") for texto in lista_modificada]
    lista_modificada = [texto.replace("Decemeber", "Dec") for texto in lista_modificada]

    lista_modificada = [texto.replace(".", "") for texto in lista_modificada]
    lista_modificada = [texto.replace(",", "") for texto in lista_modificada]

    # Lista de fechas en diferentes formatos
    fechas = lista_modificada

    # Diccionario de meses para convertir texto a números
    meses = {
        "Jan": 1, "Feb": 2, "Mar": 3, "Apr": 4, "May": 5, "Jun": 6,
        "Jul": 7, "Aug": 8, "Sep": 9, "Oct": 10, "Nov": 11, "Dec": 12
    }

    # Expresiones regulares para diferentes formatos de fecha
    patrones = [
        (r'(\d{1,2})/(\d{1,2})/(\d{4})', 'MM/DD/YYYY'),  # 03/25/1993
        (r'(\d{1,2})/(\d{1,2})/(\d{2})', 'MM/DD/YY'),  # 03/25/93

        (r'(\d{1,2})/(\d{4})', 'MM/YYYY'),  # 8/71, 7/2004
        (r'(\d{1,2})\s([A-Za-z]{3})\s(\d{2,4})', 'DD MMM YYYY'),  # 10 Sep 20
        (r'([A-Za-z]{3})\s(\d{1,2})\s(\d{2,4})', 'MMM DD YYYY'),  # Apr 11 1990
        (r'([A-Za-z]{3})\s(\d{4})', 'MMM YYYY'),  # Jul 1977,
        (r'(\d{4})', 'YYYY'),  # 2001,
        (r'(\d{1,2})/(\d{2})', 'MM/YY'),  # 25/93
    ]

    def normalizar_fecha(fecha):
        for patron, formato in patrones:
            match = re.match(patron, fecha)
            if match:
                try:
                    if formato == 'MM/DD/YY':
                        mes, dia, anio = match.groups()
                        anio = int(f"19{anio}")
                    elif formato == 'MM/DD/YYYY':
                        mes, dia, anio = match.groups()
                        anio = int(anio)
                    elif formato == 'MM/YY':
                        mes, anio = match.groups()
                        dia = 1
                        anio =int(f"19{anio}")

                    elif formato == 'MM/YYYY':
                        mes, anio = match.groups()
                        dia = 1
                        anio = int(anio)
                    elif formato == 'DD MMM YYYY':
                        dia, mes, anio = match.groups()
                        anio = int(anio) if len(anio) == 4 else int(f"20{anio}")
                        mes = meses[mes]
                    elif formato == 'MMM DD YYYY':
                        mes, dia, anio = match.groups()
                        mes = meses[mes]
                        anio = int(anio)
                    elif formato == 'MMM YYYY':
                        mes, anio = match.groups()
                        dia = 1
                        mes = meses[mes]
                        anio = int(anio)
                    elif formato == 'YYYY':
                        anio = int(match.group(1))
                        mes, dia = 1, 1


                    return datetime(anio, int(mes), int(dia))
                except ValueError:
                    return None

        return None



    mi_lista = []
    for date in fechas:
      mi_lista.append(normalizar_fecha(date))

    fechas_ordenadas = sorted(mi_lista)
    fechas_ordenadas = [fecha.strftime('%m/%d/%Y') for fecha in fechas_ordenadas]
    return fechas_ordenadas
fechas_ordenadas()
```
